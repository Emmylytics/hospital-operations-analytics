## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")
```

## 2. Load Dataset

```python
# Load the dataset into a pandas DataFrame
hospital_data = pd.read_csv('healthcare_dataset.csv')
```

## 4. Data Cleaning & Preprocessing

```python
# Remove duplicate rows to maintain data integrity
hospital_data = hospital_data.drop_duplicates()
```

```python
# Confirm that duplicates have been removed
hospital_data.duplicated().sum()
```

```python
# Verify the new dataset size after removing duplicates
hospital_data.shape
```

```python
#Standardize text formatting in the 'Name' column by converting values to title case
hospital_data['Name'] = hospital_data['Name'].str.title()
```

```python
# Preview dataset to confirm the changes
hospital_data.head()
```

```python
# Convert admission and discharge date columns to datetime format
hospital_data['Date of Admission'] = pd.to_datetime(hospital_data['Date of Admission'])
hospital_data['Discharge Date'] = pd.to_datetime(hospital_data['Discharge Date'])
```

```python
# Verify that the data types have been updated correctly
hospital_data.info()
```

## 5. Feature Engineering

```python
# Create a new feature representing the patient's length of stay in the hospital
hospital_data['Length of Stay'] = (
    hospital_data['Discharge Date'] - hospital_data['Date of Admission']
).dt.days
```

```python
# Preview the dataset to confirm the new column
hospital_data.head()
```

```python
# Add unique identifier - Patient_id
hospital_data['Patient_id'] = range(1, len(hospital_data)+1)
cols = ['Patient_id'] + [col for col in hospital_data.columns if col != 'Patient_id']
hospital_data = hospital_data[cols]
```

```python
# Rename columns to snake_case before export
hospital_data.columns = [col.lower().replace(' ', '_') for col in hospital_data.columns]
```

```python
# Preview first 5 rows of the cleaned data to verify all transformations
hospital_data.head()
```

```python
# Numeric adjustment before export
hospital_data['billing_amount'] = hospital_data['billing_amount'].round(2)
```

## 6. Export Cleaned Data

```python
# Export the cleaned dataset for further analysis in SQL and Power BI
hospital_data.to_csv('cleaned_hospital_data.csv', index=False)
```
